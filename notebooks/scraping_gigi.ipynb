```python
driver = webdriver.Chrome()
url_parafina = "https://gigistudios.com/collections/gafas-sol"
driver.get(url_parafina)
driver.maximize_window()
sleep(1)

for i in range(10):
    driver.execute_script("window.scrollTo(0, 10000);")
    sleep(3)

html_table_page = driver.page_source
driver.quit()

html_table_page
```

```python
dic_gigi={
    "nombre":[],
    "precio":[]
}

sopa = BeautifulSoup(html_table_page, "html.parser")
grid = sopa.find("ul", {"id":"product-grid"})
prorductos =  grid.findAll("li")
for producto in prorductos:
    card_content = producto.find("div", {"class":"card__content"})
  
    # Extraer el nombre del producto
    nombre = card_content.find('h3', class_='card__heading').text.strip()

    # Extraer el precio
    precio = card_content.find('span', class_='price-item--regular').text.strip()
    precio_splited = precio.split()
    
    if len(precio_splited) == 1:
        dic_gigi["nombre"].append(nombre)
        dic_gigi["precio"].append(precio.replace(",", "."))

    # Imprimir los resultados
    # print("Nombre del producto:", nombre)
    # print("Precio del producto:", precio)
```

```python
df_gigi = pd.DataFrame(dic_gigi)
df_gigi["precio"] = df_gigi["precio"].astype(float)
df_gigi.to_csv("../datos/df_gigi.csv")
```
